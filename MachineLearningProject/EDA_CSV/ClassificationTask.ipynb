```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import confusion_matrix,f1_score,accuracy_score,roc_auc_score,classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
data=pd.read_csv("https://raw.githubusercontent.com/Ravinder1191/CAPSTONE-PROJECT/main/MachineLearningProject/EDA_CSV/updated1_survey.csv")
```

```python
features=['work_interfere','family_history']
```

```python
for column in features:
       encoder=LabelEncoder()
       data[column]=encoder.fit_transform(data[column])
```

```python
label_encoder=LabelEncoder()
data['treatment']=label_encoder.fit_transform(data['treatment'])
x=data[features]
target=data['treatment']
x_train,x_test,target_train,target_test=train_test_split(x,target,test_size=0.5,random_state=42)
model=GradientBoostingClassifier()
model.fit(x_train,target_train)
predictions=model.predict(x_test)
```

```python
print(confusion_matrix(target_test,predictions))
print(f"accuracy:{accuracy_score(target_test,predictions)}")
print(f"roc auc:{roc_auc_score(target_test,predictions)}")
print(f"f1 score:{f1_score(target_test,predictions)}")
print(classification_report(target_test,predictions))
```

```python
label_encoder=LabelEncoder()
data['treatment']=label_encoder.fit_transform(data['treatment'])
x=data[features]
target=data['treatment']
x_train,x_test,target_train,target_test=train_test_split(x,target,test_size=0.5,random_state=42)
model=DecisionTreeClassifier()
model.fit(x_train,target_train)
predictions=model.predict(x_test)
```

```python
print(confusion_matrix(target_test,predictions))
print(f"accuracy:{accuracy_score(target_test,predictions)}")
print(f"roc auc:{roc_auc_score(target_test,predictions)}")
print(f"f1 score:{f1_score(target_test,predictions)}")
print(classification_report(target_test,predictions))
```
