```python
def clean_gender(gender):
    gender = str(gender).strip().lower()
    if gender.startswith("m"):
        return "Male"
    elif gender.startswith("f"):
        return "Female"
    else:
        return "Other"
data["Gender"]=data["Gender"].apply(clean_gender)
print(data["Gender"].value_counts())
```

```python
print(data.isnull().sum())
# Handling of columns having null values
data['state']=data['state'].fillna("Not Verified")
```

```python
data['self_employed']=data['self_employed'].fillna("Not Verified employed")
data['work_interfere']=data['work_interfere'].fillna("Not Verified")
data.isnull().sum()
```

```python
# using heatmap to identify features
from sklearn.preprocessing import LabelEncoder
data=pd.read_csv(r"C:\Users\Ravin\Downloads\updated1_survey.csv")
# Dropping useless column
data.drop(['Unnamed: 0','comments'],axis=1,inplace=True)
features_to_convert=['Age','Gender','Country','state', 'self_employed','family_history','treatment','work_interfere', 'no_employees','remote_work','tech_company','benefits', 'care_options']
encoder=LabelEncoder()
for feature in features_to_convert:
    data[feature]=encoder.fit_transform(data[feature])
```

```python
plt.figure(figsize=(10, 6))
correlation_matrix=data.corr(numeric_only=True)
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm', linewidths=0.9, fmt=".2f")
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```
